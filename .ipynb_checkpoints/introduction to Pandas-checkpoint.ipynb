# Pandas
This is an introduction to pandas

```python
import pandas as pd
```

There are two main data types in pandas
- Series
- DataFrame

## Series data type
This is a 1-dimentional

```python
# 2 main datatypes
cars = pd.Series(["Toyota", "BMW", "Kia", "Nissan"])
cars
```

```python
# Series = 1 dimentional
colors = pd.Series(["Red", "Blue", "Green", "Black"])
colors
```

## DataFrame data type
This is 2-dimentional

```python
# DataFrame - 2 dimentional
data_set = pd.DataFrame({"Car Manufactor": cars, "Colour": colors})
data_set
```

```python
price_data = pd.Series(missing_data["Price"])
make_data = pd.Series(missing_data["Make"])

price_make_data = pd.DataFrame({"Make": make_data, "Price": price_data})
price_make_data.plot(x="Make", y="Price", kind="bar")
```
