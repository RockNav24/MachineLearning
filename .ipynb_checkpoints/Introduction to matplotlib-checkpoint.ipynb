## Introduction to Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot()
```

```python
x = [1, 2, 3, 4, 5]
y = [12, 15, 20, 30, 50]
plt.plot(x, y)
```

```python
# 1st method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
# 2nd method 
fig = plt.figure() # creates a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recomended)
fig, ax = plt.subplots()
ax.plot(x, y); # add some data
```

## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 20, 43, 50]

# 2. Setup plot
fig, ax = plt.subplots()

# 3. Plot data
ax.plot(x, y)

# 4. Customise plot
ax.set(title="Simple Plot",
       xlabel="x-axis", 
       ylabel="y-axis")

# 5. Save & show (you sace the whole figure)
fig.savefig("sample-plot.png")
```

## Making figures with Numpy arrays
We want:
* Line plot
* Scatter plot
* Bar plot 
* Histogram
* Subplots

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x ** 2);
```

```python
# Use the same data to make a scatter 
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot

fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
butter_prices = {"butter 1": 10,
                 "butter 2": 13, 
                 "butter 3": 15}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Butter Prices",
      xlabel="butter", ylabel="prices");
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
```

```python
# Make some data for histograms 
x = np.random.randn(100)
x
```

```python
# Plot the histogram
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# Plot to each different axis

ax1.plot(x, x/2);
ax2.scatter(x, x**2)
ax3.bar(x, x)
ax4.hist(x)
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# plot to each different axis
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(x, x**2);
ax[1, 0].bar(x, x*2);
ax[1, 1].hist(x);
```

### Which one should I use pyplot vs matplotlib OO method?
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advance, use the OO mehtod

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50.head()
```

```python
# Pyplot method
over_50.plot.scatter(x="age", y="chol", c="target");
```

```python
# OO method mixed with pyplot
fix, ax = plt.subplots(figsize=(10, 10))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);

ax.set_xlim([45, 100])
```

```python
# OO mehtod form scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], 
                     y=over_50["chol"], 
                     c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line 
ax.axhline(y=over_50["chol"].mean(), linestyle="--");
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True);



# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# Customise ax0
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol");

# Add legend
ax0.legend(*scatter.legend_elements(), title="Target");

# Add horizontal line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")




# Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"]);

# Customise ax1
ax1.set(title="Heart Disease and Thalach Levels", xlabel="Age", ylabel="Thalach");

# Add legend
ax1.legend(*scatter.legend_elements(), title="Target");

# Add horizontal line
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");
```
